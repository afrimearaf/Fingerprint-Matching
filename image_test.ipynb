```python
import numpy as np
import tensorflow as tf
import cv2
import os
import matplotlib.pyplot as plt
from PIL import Image
from tensorflow import keras
```

```python
model_path = "/content/drive/MyDrive/Fingerprint/FVC2000_DB4_B/CNN_BN.h5"
model = keras.models.load_model(model_path)
```

```python
img1 = cv2.imread('/content/drive/MyDrive/Fingerprint/FVC2000_DB4_B/dataset/real_data/00000.bmp', cv2.IMREAD_GRAYSCALE)

plt.imshow(img1)
plt.show()
```

```python
img2 = cv2.imread('/content/drive/MyDrive/Fingerprint/FVC2000_DB4_B/dataset/real_data/00000.bmp', cv2.IMREAD_GRAYSCALE)

plt.imshow(img2)
plt.show()
```

```python
img1 = cv2.resize(img1, (256, 256))
img1 = np.expand_dims(img1, axis=-1)
img2 = cv2.resize(img2, (256, 256))
img2 = np.expand_dims(img2, axis=-1)
```

```python
pred1 = model.predict(np.array([img1]))
pred2 = model.predict(np.array([img2]))

if np.argmax(pred1) == np.argmax(pred2):
    print("The two fingerprints match!")
else:
    print("The two fingerprints do not match.")
```
